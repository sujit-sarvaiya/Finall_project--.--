```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class StockMarketAnalyzer:

    def __init__(self):
        self.data = None

    def load_dataset(self):
        try:
            filepath = input("Enter CSV file path : ")
            self.data = pd.read_csv(filepath)
            print("\nDataset loaded successfully!\n")
        except Exception as e:
            print("Error loading file:", e)

    def explore_data(self):
        if self.data is None:
            print("Please load dataset first!")
            return
        
        print("\nFirst 5 rows:")
        print(self.data.head())

        print("\nDataset Info:")
        print(self.data.info())

        print("\nSummary Statistics:")
        print(self.data.describe())

    def dataframe_operations(self):
        if self.data is None:
            print("Load dataset first!")
            return
        
        print("\nColumns available:", list(self.data.columns))
        col = input("Enter column name to display (example: Close): ")

        if col in self.data.columns:
            print("\nSelected Column:")
            print(self.data[col].head())
        else:
            print("Invalid column!")

    def handle_missing_data(self):
        if self.data is None:
            print("Load dataset first!")
            return
        
        print("\nMissing values before cleaning:")
        print(self.data.isnull().sum())

        self.data.fillna(method="ffill", inplace=True)

        print("\nMissing values after cleaning:")
        print(self.data.isnull().sum())

    def descriptive_statistics(self):
        if self.data is None:
            print("Load dataset first!")
            return
        
        print("\nDescriptive Statistics:")
        print(self.data.describe())

    def visualize_data(self):
        if self.data is None:
            print("Load dataset first!")
            return
        
        print("\n1. Closing Price")
        print("2. Volume")
        print("3. 20-day Moving Average")
        print("4. 50-day Moving Average")

        choice = input("Select option: ")

        plt.figure(figsize=(12, 6))
        title = ""

        if choice == "1":
            sns.lineplot(data=self.data["Close"])
            title = "Closing Price"

        elif choice == "2":
            sns.lineplot(data=self.data["Volume"])
            title = "Volume"

        elif choice == "3":
            self.data["MA20"] = self.data["Close"].rolling(20).mean()
            sns.lineplot(data=self.data["MA20"])
            title = "20-Day Moving Average"

        elif choice == "4":
            self.data["MA50"] = self.data["Close"].rolling(50).mean()
            sns.lineplot(data=self.data["MA50"])
            title = "50-Day Moving Average"

        else:
            print("Invalid choice")
            return
        
        plt.title(title)
        plt.xlabel("Index")
        plt.ylabel("Value")
        plt.show()

    def save_visualization(self):
        if self.data is None:
            print("Load dataset first!")
            return

        filename = input("Enter filename to save (example: chart.png): ")

        plt.figure(figsize=(12, 6))
        sns.lineplot(data=self.data["Close"])
        plt.title("Closing Price")
        plt.savefig(filename)

        print(f"Visualization saved as {filename}")

def main():

    analyzer = StockMarketAnalyzer()

    while True:
        print("\n=== Data Analysis & Visualization")
        print("1. Load Dataset")
        print("2. Explore Data")
        print("3. Perform DataFrame Operations")
        print("4. Handle Missing Data")
        print("5. Generate Descriptive Statistics")
        print("6. Data Visualization")
        print("7. Save Visualization")
        print("8. Exit")
        print("_._._._._._._._._._._._._._._._._._._")

        choice = input("Enter your choice: ")

        if choice == "1":
            analyzer.load_dataset()
        elif choice == "2":
            analyzer.explore_data()
        elif choice == "3":
            analyzer.dataframe_operations()
        elif choice == "4":
            analyzer.handle_missing_data()
        elif choice == "5":
            analyzer.descriptive_statistics()
        elif choice == "6":
            analyzer.visualize_data()
        elif choice == "7":
            analyzer.save_visualization()
        elif choice == "8":
            print("Exiting Program, Goodbye!!")
            break
        else:
            print("Invalid Option! Try again.")

if __name__ == "__main__":
    main()
```
